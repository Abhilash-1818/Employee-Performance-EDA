##  Employee Performance Analysis

The Human Resources (HR) department seeks to better understand the factors that impact employee performance within the organization. By analyzing employee-related data, the goal is to uncover patterns and insights that can help improve training strategies, optimize workforce allocation, and boost overall organizational performance.

This project involves conducting an Exploratory Data Analysis (EDA) on employee data to:

* Identify the key attributes associated with high-performing employees (e.g., experience, number of trainings, department, location, etc.)

* Analyze the impact of experience, trainings, and departmental differences on employee KPIs.

* Understand how demographic factors (such as age and location) relate to employee performance.

* Provide actionable insights that help HR design better training programs, improve employee engagement, and develop targeted retention strategies.

```python
# How many employee joined the org in each Year
df['DoJ'] = pd.to_datetime(df['DoJ'])

df['Joining_Year'] = df['DoJ'].dt.year

joining_per_year = df['Joining_Year'].value_counts().sort_index()

plt.figure(figsize=(10,6))
sns.lineplot(x=joining_per_year.index, y=joining_per_year.values, marker='*', linewidth=2.5, color='royalblue')
plt.title('Number of Employees Joined Each Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.grid(True)
plt.xticks(joining_per_year.index) 
plt.show()
```
